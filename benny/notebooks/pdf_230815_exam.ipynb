```python
# 2.7.1

import numpy as np
import matplotlib.pyplot as plt
inv_F = lambda y: np.log(y+1)

uniform_samples = np.random.uniform(size=1000)
samples = [inv_F(uniform_sample) for uniform_sample in uniform_samples]

plt.scatter(samples, uniform_samples)
plt.show()
```

```python
# 2.7.2

mean = np.mean(samples)
var = np.var(samples)
print(mean)
print(var)
```
